##5. Employee Salary Analyzer
Problem:
Store employee names and salaries.in dictionary

Find:

highest salary,

lowest salary,

average salary.

Example:
**bold text**
employees = {
    "John":50000,
    "Ram":70000,
    "Anu":40000
}

Output:


Highest : Ram
Lowest : Anu
Average : 53333

Concepts:

dictionaries

loops

aggregation logic
( with oops)

```python
class Employees_anlysis:

  def __init__(self,employees):
    self.emp=employees
  def analyze_salaries(self):
    if not self.emp:
        print("No employee records found.")# handling empty ..
        return

    total_emp=0
    total_amount=0

    max_salary_emp=[]
    min_salary_emp=[]

    max=float('-inf')
    min=float('inf')

    for name , salary in self.emp.items():
      total_emp +=1
      total_amount += salary
      if salary >max:
        max =salary
      if salary < min:
        min =salary

    avg_salary=total_amount/total_emp

    for name ,salary in self.emp.items():
      if salary == min:
        min_salary_emp.append(name)
      elif salary == max:
        max_salary_emp.append(name)

    # Display Results
    print(f"Highest salary employees : {max_salary_emp} salary is (${max})")
    print(f"Lowest salary employees: {min_salary_emp} salary is (${min})")
    print(f"Average salary employees: {avg_salary:.0f}")

# input....
employees = {
    "John": 50000,
    "Ram": 70000,
    "Anu": 40000,
    "Sita": 65000,
    'gana':70000
}

obj=Employees_anlysis(employees) #object creating
obj.analyze_salaries()
```
